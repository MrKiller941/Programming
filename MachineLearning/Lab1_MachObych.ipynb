```python
fig, ax = plt.subplots()

ax.set_title("Closing stock value of Alphabet Inc.")
ax.set_xlabel("Date")
ax.set_ylabel("Closing Value")

numbers = [772.5, 776.4, 776.5, 776.8, 775.1]
dates = [
    datetime(2016,10,3),
    datetime(2016,10,4),
    datetime(2016,10,5),
    datetime(2016,10,6),
    datetime(2016,10,7),
]

ax.set_xlim(datetime(2016,10,3), datetime(2016,10,7))
ax.set_ylim(772.5, 777.0)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
ax.xaxis.set_major_formatter(DateFormatter("%Y-%m-%d"))
ax.xaxis.set_major_locator(mdates.DayLocator())
ax.tick_params(which='major', labelsize=10)
ax.plot(dates, numbers, c='r', linewidth=0.8)
ax.scatter(dates, numbers, edgecolor='black', color='r')
plt.show()
```

# Задание №2.

```python
X = np.linspace(0.5, 3.5, 100)
Y1 = 3 + np.cos(X)
Y2 = 1 + np.cos(1 + X / 0.75) / 2
Y3 = np.random.uniform(Y1, Y2, len(X))

fig = plt.figure(figsize=(7.5, 7.5))
ax = fig.add_axes([0.2, 0.17, 0.68, 0.7], aspect=1)

ax.xaxis.set_major_locator(MultipleLocator(1.000))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_major_locator(MultipleLocator(1.000))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

ax.tick_params(which='major', width=1.0, length=10, labelsize=14)
ax.tick_params(which='minor', width=1.0, length=5, labelsize=10,
               labelcolor='0.25')

ax.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10)

ax.plot(X, Y1, c='b', lw=2.5, label="Синяя линия", zorder=10)
ax.plot(X, Y2, c='r', lw=2.5, label="Красная линия")
ax.plot(X[::3], Y3[::3], linewidth=0, markersize=7,
        marker='o', markerfacecolor='none', markeredgecolor='black',
        markeredgewidth=2.5)

ax.set_title("Элементы изображения", fontsize=20, verticalalignment='bottom')
ax.set_xlabel("Подпись оси OX", fontsize=14)
ax.set_ylabel("Подпись оси OY", fontsize=14)
ax.legend(loc="upper right", fontsize=14)

plt.show()
```

```python
fig, ax = plt.subplots()

X = np.arange(-10, 10, 0.02)
Y = lambda X: np.log((X ** 2 + 1) * np.exp(-np.absolute(X) / 10)) / np.log(1 + np.tan(1 / (1 + np.sin(X) * np.sin(X))))

ax.plot(X, Y(X))
plt.show()
```
