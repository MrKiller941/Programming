# Задание №2.

```python
df = pd.read_csv('D:\Lab4\howpop_train.csv')
df
```

```python
df.drop(filter(lambda c: c.endswith("_lognorm"), df.columns), axis=1, inplace=True,) 
df
```

```python
df["published"] = pd.to_datetime(df.published, yearfirst=True)
df.dtypes
```

```python
df["year"] = [d.year for d in df.published]
df["month"] = [d.month for d in df.published]
df["dayofweek"] = [d.isoweekday() for d in df.published]
df["hour"] = [d.hour for d in df.published]
df
```

```python
df_t = df.astype({'year': str, 'month': str})
df["year-month"] = df_t['year'] + "-" + df_t['month']
ax = sb.countplot(x=df["year-month"])
months = df.groupby(df["year-month"])["year-month"].count().shape[0]
ax.set(xticks=(range(0, months, 12)))
plt.show()
```

# Больше всего публикаций в интервале 2015 - 3.
