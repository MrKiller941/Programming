```python
def tribonacci(n):
    if (n==2): return 1     # базовый случай
    elif (n < 2): return 0  # базовый случай
    else:
        return tribonacci(n-1)+tribonacci(n-2)+tribonacci(n-3) # рекурсивный случай
print("Введите номер числа Трибоначчи")
n = int(input())
print(tribonacci(n))    
```
