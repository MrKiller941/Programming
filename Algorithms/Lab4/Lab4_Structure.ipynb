```python
class HashTable:
    def __init__(self):
        self.size = 11
        self.len = 0
        self.step = 1
        self.slots = [None] * self.size
        self.data = [None] * self.size

    def put(self, key, data):

        nextslot = self.hashfunction(key, len(self.slots))
        while self.slots[nextslot] != None and \
                self.slots[nextslot] != key:
            nextslot = self.hashfunction(nextslot, len(self.slots), True)

        if self.slots[nextslot] == None:
            self.slots[nextslot] = key
            self.data[nextslot] = data
        else:
            self.data[nextslot] = data
            
        self.step = 1
        
        self.relen()
        
        if ((self.size - self.len) / self.size < 0.3):
            self.resize()
            
    def hashfunction(self, key, size, Next = False):
        if(not Next):
            return key % size
        else:
            return (key + self.nextstep()) % size
    
    def resize(self, increase = False):
        oldsize = self.size
        
        if (increase and self.size/2 <= 11):
            self.size = 11
        else:
            lst = self.prime_number(self.size*2 if not increase else self.size/2)
            self.size = lst[-2] if self.size - lst[-2] <= lst[-1] - self.size else lst[-1]

        if increase:
            oldslots = self.slots
            olddata = self.data
            self.slots = [None] * self.size
            self.data = [None] * self.size
            for key,data in zip(oldslots, olddata):
                if key != None and data != None:
                    self.put(key,data)
        else:
            newslots = [None] * self.size
            newdata = [None] * self.size
            newslots[0:oldsize] = self.slots[0:oldsize]
            newdata[0:oldsize] = self.data[0:oldsize]
            self.slots = newslots
            self.data = newdata
        
            
    def prime_number(self, n):
        lst=[2]
        end = False
        i = 3
        while (lst[-1] <= n):
            if (i > 10) and (i%10==5):
                i += 2
                continue
            for j in lst:
                if j*j-1 > i:
                    lst.append(i)
                    break
                if (i % j == 0):
                    break
            else:
                
                lst.append(i)
            i += 2
        return lst


    def nextstep(self):
        step = self.step**2
        self.step += 1
        return step
        
    def get(self, key):
        startslot = self.hashfunction(key, len(self.slots))

        data = None
        stop = False
        found = False
        position = startslot
        while self.slots[position] != None and \
                not found and not stop:
            if self.slots[position] == key:
                found = True
                data = self.data[position]
            else:
                position = self.hashfunction(position, len(self.slots), True)
                if position == startslot:
                    stop = True
        return data
        
    def relen(self):
        size = 0
        for x in self.data:
            if x:
                size += 1
        self.len = size
        
    def __getitem__(self, key):
        return self.get(key)

    def __setitem__(self, key, data):
        self.put(key, data)
        
    def __len__(self):
        return self.len
    
    def __contains__(self, item):
        return item in self.data
    
    def __delitem__(self, item):
        if item in self.slots:
            index = self.slots.index(item)
            self.slots[index] = None
            self.data[index] = None
        
            self.relen()
        
            if ((self.size - self.len) / self.size > 0.8):
                self.resize(True)
```

```python
class HashTable:
    def __init__(self):
        self.size = 13
        self.len = 0
        self.step = 1
        self.slots = [None] * self.size
        self.data = [None] * self.size

    def put(self, key, data):
        nextslot = self.hashfunction(key, len(self.slots))
        
        while True:         
            if self.slots[nextslot] == None:
                self.slots[nextslot] = UnorderedList()
                self.data[nextslot] = UnorderedList()
                self.slots[nextslot].append(key)
                self.data[nextslot].append(data)
                break;
            elif self.slots[nextslot].size() < 3:
                if self.slots[nextslot].search(key):
                    self.data[nextslot][self.slots[nextslot].index(key)] = data
                    break;
                else:
                    self.slots[nextslot].append(key)
                    self.data[nextslot].append(data)
                    break;

            nextslot = self.hashfunction(nextslot, len(self.slots), True)
        self.relen()
        if ((self.size - self.number_none()) / self.size < 0.3):
            self.resize()
            
    def hashfunction(self, key, size, Next = False):
        if (type(key) != int):
            key = int.from_bytes(key.encode('utf-8'), 'little')
        if(not Next):
            return key % size
        else:
            return (key + self.nextstep()) % size
            
    def resize(self, increase = False):
        oldsize = self.size
        
        if (increase and self.size/2 <= 11):
            self.size = 11
        else:
            lst = self.prime_number(self.size*2 if not increase else self.size/2)
            self.size = lst[-2] if self.size - lst[-2] <= lst[-1] - self.size else lst[-1]
        
        if increase:
            oldslots = self.slots
            olddata = self.data
            self.slots = [None] * self.size
            self.data = [None] * self.size
            for key,data in zip(oldslots, olddata):
                if key != None and data != None:
                    for x,y in zip(key, data):
                        self.put(x.getData(),y.getData())
        else:
            newslots = [None] * self.size
            newdata = [None] * self.size
            newslots[0:oldsize] = self.slots[0:oldsize]
            newdata[0:oldsize] = self.data[0:oldsize]
            self.slots = newslots
            self.data = newdata
            
    def prime_number(self, n):
        lst=[2]
        end = False
        i = 3
        while (lst[-1] <= n):
            if (i > 10) and (i%10==5):
                i += 2
                continue
            for j in lst:
                if j*j-1 > i:
                    lst.append(i)
                    break
                if (i % j == 0):
                    break
            else:
                lst.append(i)
            i += 2
        return lst

    def nextstep(self):
        step = self.step**2
        self.step += 1
        return step
        
    def get(self, key):
        startslot = self.hashfunction(key, len(self.slots))

        data = None
        stop = False
        found = False
        position = startslot
        while self.slots[position] != None and \
                not found and not stop:
            if self.slots[position].search(key):
                found = True
                data = self.data[position][self.slots[position].index(key)]
            else:
                position = self.hashfunction(position, len(self.slots), True)
                if position == startslot:
                    stop = True
        return data

    def number_none(self):
        number = 0
        for i in self.slots:
            if i:
                number += 1
        return number
        
    def relen(self):
        size = 0
        for x in self.data:
            if x:
                size += x.size()
        self.len = size
        
    def __getitem__(self, key):
        return self.get(key)

    def __setitem__(self, key, data):
        self.put(key, data)
        
    def __len__(self):
        return self.len
    
    def __contains__(self, item):
        find = False
        for i in self.data:
            if i:
                if i.search(item):
                    find = True
        return find
    
    def __delitem__(self, item):
        
        for i, (key, data) in enumerate(zip(self.slots,self.data)):
            if key:
                if key.search(item):
                    index = key.index(item)
                    key.pop(index)
                    data.pop(index)
                    if key.isEmpty():
                        self.slots[i] = None
                        self.data[i] = None
                    self.relen()
                    if ((self.size - self.number_none()) / self.size > 0.8):
                        self.resize(True)
                    return
```
